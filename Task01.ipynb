```python
#first we import the necessary files

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# we have use skiprows to skip the first 4 rows of the dataset as they are not needed

df = pd.read_csv(r"API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv", skiprows = 4)
df.head()
```

```python
print("Columns Given in data:", df.columns)
```

```python
df.describe()
```

```python
#  Now, we are creating Bar chart , Histogram for the given data.

# 1. Bar chart  - Let us select Year 2023.


missing_count = df['2023'].isna().sum()
print(f"Number of missing values in 2023: {missing_count}")

# Clean data
df_cleaned = df.dropna(subset=['2023'])
df_cleaned.loc[:, '2022'] = df_cleaned['2023'].astype(float)

plt.figure(figsize=(60, 12))
plt.bar(df_cleaned['Country Name'], df_cleaned['2023'], color='skyblue')
plt.title("Total Population in 2023 (All Countries)", fontsize=20)
plt.xlabel("Country", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.xticks(rotation=90, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
top_20_countries = df_cleaned.sort_values(by='2023', ascending=False).head(20)

plt.figure(figsize=(10, 6))
plt.bar(top_20_countries['Country Name'], top_20_countries['2023'], color='skyblue')
plt.title("Top 20 Countries by Population in 2023", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#  In this I try to create TWO BAR CHARTS for the year 2012 and 2022 together with is interesting part to learn here.
#  Then, i have selected some countries to show the population in 2012 and 2022.

df_cleaned = df.dropna(subset=['2012', '2022'])
df_cleaned.loc['2012'] = df_cleaned['2012'].astype(float)
df_cleaned.loc['2022'] = df_cleaned['2022'].astype(float)

selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Nigeria']
selected_df = df_cleaned[df_cleaned['Country Name'].isin(selected_countries)]

countries = selected_df['Country Name']
pop_2012 = selected_df['2012']
pop_2022 = selected_df['2022']


import numpy as np
x = np.arange(len(countries))
bar_width = 0.35

# Plot side-by-side bars  --> New Interesting part to learn
plt.figure(figsize=(12, 6))
plt.bar(x - bar_width/2, pop_2012, width=bar_width, label='2012', color='orange')
plt.bar(x + bar_width/2, pop_2022, width=bar_width, label='2022', color='skyblue')


plt.title("Population Comparison (2012 vs 2022)", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(x, countries, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram - For Year 2023

year = '2023'
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned[year], bins=30, color='orange', edgecolor='black')
plt.title(f"Distribution of Population Across Countries in {year}")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
year = '2023'

top20 = df_cleaned[['Country Name', year]].sort_values(by=year, ascending=False).head(20)

plt.figure(figsize=(12, 8))
plt.barh(top20['Country Name'], top20[year], color='skyblue', edgecolor='black')
plt.title(f"Top 20 Most Populous Countries in {year}")
plt.xlabel("Population")
plt.ylabel("Country")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
